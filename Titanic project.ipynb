```python
x=data.drop('Survived',axis=1).values
y=data['Survived'].values
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Fill missing values with the mean of each column
imputer = SimpleImputer(strategy='mean')
x_train_imputed = imputer.fit_transform(x_train)
x_test_imputed = imputer.transform(x_test)

# Train the logistic regression model
lr = LogisticRegression(max_iter=1000)
lr.fit(x_train_imputed, y_train)

# Make predictions
lr_pred = lr.predict(x_test_imputed)

# Optional: Check performance
print("Accuracy:", accuracy_score(y_test, lr_pred))
```

```python
from sklearn.model_selection import GridSearchCV
import numpy as np

# Define parameter space
C_space = np.logspace(-5, 8, 15)
param_grid = {'C': C_space}

# Create GridSearchCV object
logreg_cv = GridSearchCV(LogisticRegression(max_iter=1000), param_grid, cv=5)

# Fit model
logreg_cv.fit(x_train_imputed, y_train)

# Print the best parameters and score
print("Tuned Logistic Regression Parameters: {}".format(logreg_cv.best_params_))
print("Best score is: {:.4f}".format(logreg_cv.best_score_))
```
